```python
from l2cs import Pipeline, render
import cv2
import torch

# 1. 모델 파이프라인 초기화
gaze_pipeline = Pipeline(
    weights='models/L2CSNet_gaze360.pkl',
    arch='ResNet50',
    device=torch.device('cuda') # or 'gpu'
)
 
# 2. 웹캠 설정
# cam 변수를 정의해야 합니다. 0은 기본 웹캠을 의미합니다.
video_id = 'webcam.mp4'
cap = cv2.VideoCapture(video_id)
if not cap.isOpened():
    print(f"웹캠 {video_id}을 열 수 없습니다.")
    exit()

# 3. 실시간 처리 루프
while True:
    # 프레임 읽기
    success, frame = cap.read()
    if not success:
        print("프레임을 읽는 데 실패했습니다.")
        break
    frame = cv2.resize(frame, (960, 540))
    
    # L2CS-Net 파이프라인으로 시선 예측
    results = gaze_pipeline.step(frame)
    # 예측 결과를 프레임에 시각화
    # render 함수는 얼굴 바운딩 박스와 시선 벡터를 그려줍니다.
    frame = render(frame, results)

    # 화면에 출력
    cv2.imshow("L2CS-Net Demo", frame)

    # 'q' 키를 누르면 종료
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# 4. 리소스 해제
cap.release()
cv2.destroyAllWindows()
```
